```python
# load packages
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
base_site='https://www.autotrader.co.za/cars-for-sale/ford/mustang?sortorder=PriceHigh&price=less-than-2000000'
```

```python
def get_data(url):
    response=requests.get(url)
    #print(response.status_code)
    html=response.content
    data=BeautifulSoup(html,'lxml')
    return data
```

```python
get_data(base_site)
```

```python
def get_next_page(data):
    pager=data.find('div',{'class':'b-pagination-bar'})
    if not pager.find('a',{'class':'gm-float-right m-disabled e-pagination-link'}):
        url=pager.find('a',{'class':'gm-float-right e-pagination-link'})['href']
        return 'https://www.autotrader.co.za'+url
    else:
        return
```

```python
url=base_site
while True:
    time.sleep(1)
    data=get_data(url)
    url=get_next_page(data)
    if not url:
        break
    print(url)
```

```python

def string_manipulation(icons_data):
    #Removing any space at the start or end of the string plus the values used for indentation 
    icons_data=icons_data.strip()
    icons_data=icons_data.replace('\xa0','')
    
    t_year=icons_data[:4]
    t_trans=''
    t_km=''
    
    temp_data=icons_data[4:]
    if temp_data[-9:]=='Automatic':
        t_trans=temp_data[-9:]
        t_km=temp_data.replace(temp_data[-9:],'')
        
    elif temp_data[-6:]=='Manual':
        t_trans=temp_data[-6:]
        t_km=temp_data.replace(temp_data[-9:],'')
    else:
        t_trans='Not Specified'
        
    #km.append(temp_data)
    return [t_year,t_trans,t_km]
    
string_manipulation(original)
    
```

```python
price=[]
#location=[]
name=[]
year=[]
kilometers=[]
transmission=[]
dealer=[]

def insert_data(data):
    car_list=data.find_all('div',class_='e-available m-has-photos')

    for index in range(len(car_list)):
        #storing the information about every propery 
        tempPrice=car_list[index].find('span',{'class':'e-details','class':'e-price-container'}).text.strip()
        tempName=car_list[index].find('span',{'class':'e-details','class':'e-title'}).text.strip()
        tempDealer=car_list[index].find('span',{'class':'e-details','class':'e-dealer-details'}).text.strip()
        icons_data=car_list[index].find('span',{'class':'e-details','class':'e-icons'}).text.strip()
        icon_list=string_manipulation(icons_data)
        

        #loading the data to lists
        price.append(tempPrice)
        name.append(tempName)
        dealer.append(tempDealer)
        year.append(icon_list[0])
        transmission.append(icon_list[1])
        kilometers.append(icon_list[2])
```

```python
url=base_site
while True:
    time.sleep(1)
    data=get_data(url)
    insert_data(data)
    url=get_next_page(data)
    if not url:
        break
    print(url)
```

```python
AutoTrader=pd.DataFrame()

AutoTrader['Name']=name
AutoTrader['Price']=price
AutoTrader['Dealer']=dealer
AutoTrader['Year']=year
AutoTrader['Transmission']=transmission
AutoTrader['KM']=kilometers
```
